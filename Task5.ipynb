Group by function

Lets suppose we want to see the number of SALES according to PRODUCTLINE

```python
sales_by_productline = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
```

```python
print('Sales according to Productline:\n', sales_by_productline)
```

# Plotting

Plotting a bar chart visualizing sales with product line

```python
import matplotlib.pyplot as plt

#Creating a bar chart from ur grouped data

sales_by_productline.plot(kind = 'bar', figsize=(10,6), title = 'Total Sales by Productline')

# Adding labels
plt.ylabel('Total Sales (in millions)')
plt.xlabel('Product Line')

# Displaying the chart
plt.show()
```
